```python
from sklearn.model_selection import train_test_split 
```

```python
X = users[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = users['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
predicts = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predicts))
print(confusion_matrix(y_test,predicts))
```

# Conclusion

Based on the test results we can conclude that our model have a good accuracy rate with room for improvement but a 91% accuracy rate is already impressive.
